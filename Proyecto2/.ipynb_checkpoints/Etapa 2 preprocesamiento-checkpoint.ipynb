## Carga de librerías

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
# Para preparar los datos
from sklearn.preprocessing import LabelEncoder
```

Carga de los archivos

```python
datos2017= pd.read_csv(r'C:\Users\Camila\Documents\Universidad\Septimo semestre\BI\Proyecto\Proyecto 2 etapa 2\data_2010_2017.csv', sep=',', encoding = 'utf-8')
datos2017.info()
```

```python
datos2018= pd.read_csv(r'C:\Users\Camila\data_2018_2020.csv', sep=',', encoding = 'utf-8')
datos2018.info()
```

```python
#Eliminación "TOTAL"
#Datos 2017
for col in datos2017.columns:
    datos2017.drop(datos2017.loc[datos2017[col]=='TOTAL'].index,inplace=True)
#Datos 2018-2020
for col in datos2018.columns:
    datos2018.drop(datos2018.loc[datos2018[col]=='TOTAL'].index,inplace=True)
```

```python
#Verificación
datos2017.head(50)
```

```python
#Verificación
datos2018.head(50)
```

```python
datos2017.shape
```

```python
datos2018.shape
```

```python
#Juntar dataframe
# Apilar los __DataFrames__ uno encima del otro
datoscompletos = pd.concat([datos2017, datos2018], axis=0)
```

```python
#Verificación paso anterior
datoscompletos.head(50)
```

```python
print(datoscompletos.shape)
```

```python
#Manejo extranjeros
#¿Cuántos extranjeros?
for cols in datoscompletos:
    print(datoscompletos[cols].value_counts())
```

```python
#Extranjeros
datoscompletos["codigo_depto"]=datoscompletos.codigo_depto.replace({'Extranjeros':1001})
datoscompletos["codigo_muni"]=datoscompletos.codigo_muni.replace({'Extranjeros':1001})
```

```python
#Pasar los datos que no son numéricos a si
datoscompletos = datoscompletos.astype({"codigo_depto":int,"codigo_muni":int,"codigo_evento":int})
datoscompletos.info()
```

```python
datoscompletos.head(20)
```

## Normalización de datos

Se obtuvieron los datos de Terradata acerca de las poblaciones históricas.

```python
#Carga de datos
poblaciones= pd.read_excel(r'C:\Users\Camila\Documents\Universidad\Septimo semestre\BI\Proyecto\Proyecto 2 etapa 2\TerriData_Dim25_Sub10_Var1\TerriData_Dim25_Sub10_Var1.xlsx')
poblaciones.info()
```

```python
poblaciones.head(50)
```

```python
#Eliminar columnas innecesarias
poblaciones.drop(['Dimensión', 'Subcategoría', 'Indicador', 'Dato Cualitativo', 'Mes', 'Fuente', 'Unidad de Medida'], axis=1, inplace=True)
```

```python
#Manejo columnas
poblaciones.columns = ['codigo_depto', 'nombre_depto', 'codigo_muni', 'nombre_muni', 'poblacion_municipio', 'anio']
poblaciones.head(20)
```

```python
#Merge
datosnormalizados = pd.merge(datoscompletos, poblaciones,on=['codigo_muni', 'codigo_depto', 'anio', 'nombre_muni', 'nombre_depto'])
```

```python
datosnormalizados.head(645689)
```

```python
datosnormalizados.shape
```

```python
#Añadir extranjeros
extranjeros=datoscompletos['nombre_depto']=='Extranjeros'
positions = np.flatnonzero(extranjeros)
filtros=datoscompletos.iloc[positions]
filtros.head(10)
```

```python
datosnormalizados.append(filtros)
```

```python
datosnormalizados['caso-poblacion']= (datosnormalizados['num_casos']/datosnormalizados['poblacion_municipio'])*100000
```

```python
datosnormalizados.head(50)
```

## Estadísticas


```python
#Por region
datosnormalizados.groupby(['nombre_depto']).sum()['caso-poblacion'].plot(kind = 'bar', figsize=(15, 6))
```

```python
#Por género
datosnormalizados.groupby(['sexo']).sum()['caso-poblacion'].plot(kind = 'bar', figsize=(15, 6))
```

```python
#Por evento
datosnormalizados.groupby(['evento']).sum()['caso-poblacion'].plot(kind = 'line', figsize=(45, 15))
```

```python
# Resumen de los estadisticos de las variables numéricas
datosnormalizados.describe()
```

## Para ETL

```python
#Genero
genero = pd.DataFrame()
genero['descripcion'] = datosnormalizados['sexo'].unique()
genero
```

```python
#Ubicacion
ubicacion = pd.DataFrame()
ubicacion['anio'] = datosnormalizados['anio']
ubicacion['nombre_depto'] = datosnormalizados['nombre_depto']
ubicacion['codigo_depto']=datosnormalizados['codigo_depto']
ubicacion['nombre_muni']=datosnormalizados['nombre_muni']
ubicacion['codigo_muni']=datosnormalizados['codigo_muni']
ubicacion['poblacion_muni']=datosnormalizados['poblacion_municipio']
```

```python
ubicacion = ubicacion.drop_duplicates()
ubicacion.head(20)
```

```python
#Causa
causa = pd.DataFrame()
causa['nombre'] = datosnormalizados['evento'].unique()
causa
```

```python
#Fecha
fecha = pd.DataFrame()
fecha['fecha'] = datosnormalizados['anio'].unique()
fecha
```

```python
#Edad
edad = pd.DataFrame()
edad['edad'] = datosnormalizados['edad'].unique()
edad
```

## Extracción de datos

```python
#Ubicacion
ubicacion.to_csv("ubicacion.csv", encoding='utf-8', index=False)
```
